```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data = pd.read_csv('HRDataset_v14.csv')
data.head()
```

```python
sns.countplot(x=data.Sex)
```

```python
sns.countplot(x=data.MaritalDesc, hue=data.Sex)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=data.PerformanceScore, hue=data.MaritalDesc)
plt.legend(loc='upper right');
```

```python
sns.countplot(y=data['RaceDesc'], order=data['RaceDesc'].value_counts().index);
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Department", y="Salary", data=data,hue="Sex", ci=None, estimator=np.mean);
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Department", y="EmpSatisfaction", data=data, ci=None, estimator=np.mean, order=data.groupby(["Department"])['EmpSatisfaction'].mean().sort_values(ascending=False).index);
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="ManagerName", y="EmpSatisfaction", data=data, ci=None, estimator=np.mean, order=data.groupby(["ManagerName"])['EmpSatisfaction'].mean().sort_values(ascending=False).index)
plt.xticks(rotation =90);
```

```python
order_position_salary = data.groupby(["Position"])['Salary'].mean().sort_values(ascending=False)
```

```python
fig, axes = plt.subplots(2, sharex=True, figsize=(15,10))
sns.barplot(ax=axes[0], x='Position', y='Salary', data=data, ci=None, estimator=np.mean, order=order_position_salary.index)
sns.barplot(ax=axes[1], x='Position', y='EmpSatisfaction', data=data, ci=None, estimator=np.mean, order=order_position_salary.index)
plt.xticks(rotation = 90);
```
